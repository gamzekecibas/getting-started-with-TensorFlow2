- **Step 01.** Import libraries 

```python
import tensorflow as tf
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
import random

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, Conv2D, MaxPooling2D, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.optimizers import Adam
```

- **Step 02.** Load and preprocess data

```python
# Run this cell to load the dataset

train = loadmat('train_32x32.mat')
test = loadmat('test_32x32.mat')
```

```python
train_data= train['X']
train_label= train['y']
test_data= test['X']
test_label= test['y']

train_data = np.moveaxis(train_data, -1, 0)
test_data = np.moveaxis(test_data, -1, 0)
train_label = np.where(train_label==10, 0, train_label)
test_label = np.where(test_label==10, 0, test_label)

print(train_data.shape)
print(test_data.shape)
```

```python
fig_rnd= np.random.randint(0,len(train_data),10)

for ind in fig_rnd:
    train_fig= np.array(train_data[ind,:,:,:])
    plt.figure()
    plt.imshow(train_fig)
```

```python
# Gray scale
gtrain_data= np.mean(train_data, -1, keepdims=False)
gtest_data= np.mean(test_data, -1, keepdims=False)

for ind in fig_rnd:
    train_fig= np.array(gtrain_data[ind,:,:])
    plt.figure()
    plt.imshow(train_fig)
```

- **Step 03.** Model Investigation
    - **a. MLP** neural network classifier
* Build an MLP classifier model using the Sequential API. Your model should use only Flatten and Dense layers, with the final layer having a 10-way softmax output. 
* You should design and build the model yourself. Feel free to experiment with different MLP architectures. _Hint: to achieve a reasonable accuracy you won't need to use more than 4 or 5 layers._

```python
model_mlp = Sequential([ Flatten(input_shape=train_data[0].shape), Dense(256, activation='relu'),
                        Dense(128, activation='relu'), Dense(128, activation='relu'),
                        Dense(10, activation='softmax')])
model_mlp.summary()
```

> - **a. MLP** neural network classifier
>> **1. Callbacks creating**

```python
cp_path = 'Checkpoint/Best_model'
cp = ModelCheckpoint(cp_path,
                     save_best_only=True,
                     save_weights_only=True,
                     verbose=False,
                     save_freq='epoch',
                     monitor='val_loss',
                     mode='min')
EarlyStop = EarlyStopping(monitor='val_loss', mode='min', patience=2)
```

> - **a. MLP** neural network classifier
>> **2. Model compiling**

```python
model_mlp.compile(optimizer= 'adam', loss=tf.keras.losses.sparse_categorical_crossentropy, metrics= ['accuracy'])
```

> - **a. MLP** neural network classifier
>> **3. Model training**

```python
history_mlp= model_mlp.fit(train_data, train_label, epochs=20, batch_size=32,
                           verbose=True, validation_split=0.2, callbacks=[cp, EarlyStop])
```

> - **a. MLP** neural network classifier
>> **4. Learning plots**

```python
plt.plot(history_mlp.history['loss'],'b.')
plt.plot(history_mlp.history['val_loss'],'b-')
plt.title("Loss for MLP Model")
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['Training Loss', 'Validation Loss'])
plt.show()
```

```python
plt.plot(history_mlp.history['accuracy'],'r.')
plt.plot(history_mlp.history['val_accuracy'],'r-')
plt.title("Accuracy for MLP")
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['Training Accuracy', 'Validation Accuracy'])
plt.show()
```

> - **a. MLP** neural network classifier
>> **5. Model evaluation**

```python
test_loss_mlp, test_accuracy_mlp = model_mlp.evaluate(test_data, test_label, verbose=False)
print(f"Test Loss is {test_loss_mlp} for MLP.")
print(f"Test Accuracy is {test_accuracy_mlp} for MLP.")
```

> - **b. CNN** neural network classifier
* Build a CNN classifier model using the Sequential API. Your model should use the Conv2D, MaxPool2D, BatchNormalization, Flatten, Dense and Dropout layers. The final layer should again have a 10-way softmax output. 
* You should design and build the model yourself. Feel free to experiment with different CNN architectures. _Hint: to achieve a reasonable accuracy you won't need to use more than 2 or 3 convolutional layers and 2 fully connected layers.)_
* The CNN model should use fewer trainable parameters than your MLP model.

```python
model_cnn = Sequential([ 
    Conv2D(16, (3,3), padding='same', activation='relu', input_shape=train_data[0].shape),
    MaxPooling2D((2,2)), 
    Conv2D(16, (3,3), activation='relu'),
    MaxPooling2D((2,2)), 
    Conv2D(16, (3,3), padding='same', activation='relu'),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    Dense(10, activation='softmax')]
)
model_cnn.summary()
```

> - **b. CNN** neural network classifier
>> **1. Callbacks creating**

```python
cp_path_cnn = 'CNN_Checkpoint/Best_model'
cp_cnn = ModelCheckpoint(cp_path_cnn,
                     save_best_only=True,
                     save_weights_only=True,
                     verbose=False,
                     save_freq='epoch',
                     monitor='val_loss',
                     mode='min')
EarlyStop_cnn = EarlyStopping(monitor='val_loss', mode='min', patience=2)
CB_cnn= [cp_cnn, EarlyStop_cnn]
```

> - **b. CNN** neural network classifier
>> **2. Model compiling**

```python
model_cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

> - **b. CNN** neural network classifier
>> **3. Model training**

```python
history_cnn = model_cnn.fit(train_data, train_label, epochs=20, batch_size=64,
                        verbose=True, validation_split=0.2, callbacks=CB_cnn)
```

> - **b. CNN** neural network classifier
>> **4. Learning plots**

```python
plt.figure()
plt.plot(history_cnn.history['loss'],'b.')
plt.plot(history_cnn.history['val_loss'], 'b-')
plt.title("Loss of CNN")
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['Training Loss', 'Validation Loss'])
plt.show()

plt.figure()
plt.plot(history_cnn.history['accuracy'],'r.')
plt.plot(history_cnn.history['val_accuracy'],'r-')
plt.title("Accuracy of CNN")
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['Training Accuracy', 'Validation Accuracy'])
plt.show()
```

> - **b. CNN** neural network classifier
>> **5. Model evaluation**

```python
test_loss_cnn, test_accuracy_cnn = model_cnn.evaluate(test_data, test_label, verbose=False)
print(f"Test Loss is {test_loss_cnn}")
print(f"Test Accuracy is {test_accuracy_cnn}")
```

- **Step 05.** Making prediction

```python
num_test_images = test_data.shape[0]

random_inx = np.random.choice(num_test_images, 5)
random_test_images = test_data[random_inx, ...]
random_test_labels = test_label[random_inx, ...]

predictions = model_mlp.predict(random_test_images)

fig, axes = plt.subplots(5, 2, figsize=(16, 12))
fig.subplots_adjust(hspace=0.4, wspace=-0.2)

for i, (prediction, image, label) in enumerate(zip(predictions, random_test_images, random_test_labels)):
    axes[i, 0].imshow(np.squeeze((image * 255).astype(np.uint8)))
    axes[i, 0].get_xaxis().set_visible(False)
    axes[i, 0].get_yaxis().set_visible(False)
    axes[i, 0].text(10., -1.5, f'Digit {label}')
    axes[i, 1].bar(np.arange(len(prediction)), prediction)
    axes[i, 1].set_xticks(np.arange(len(prediction)))
    axes[i, 1].set_title(f"Categorical distribution. Model prediction: {np.argmax(prediction)}")
    
plt.show()
```

```python
# For CNN
num_test_images = test_data.shape[0]

random_inx = np.random.choice(num_test_images, 5)
random_test_images = test_data[random_inx, ...]
random_test_labels = test_label[random_inx, ...]

predictions = model_cnn.predict(random_test_images)

fig, axes = plt.subplots(5, 2, figsize=(16, 12))
fig.subplots_adjust(hspace=0.4, wspace=-0.2)

for i, (prediction, image, label) in enumerate(zip(predictions, random_test_images, random_test_labels)):
    axes[i, 0].imshow(np.squeeze((image * 255).astype(np.uint8)))
    axes[i, 0].get_xaxis().set_visible(False)
    axes[i, 0].get_yaxis().set_visible(False)
    axes[i, 0].text(10., -1.5, f'Digit {label}')
    axes[i, 1].bar(np.arange(len(prediction)), prediction)
    axes[i, 1].set_xticks(np.arange(len(prediction)))
    axes[i, 1].set_title(f"Categorical distribution. Model prediction: {np.argmax(prediction)}")
    
plt.show()
```
